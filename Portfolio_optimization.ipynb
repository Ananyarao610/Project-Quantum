```python
dates = pd.date_range('2000-01-01', '2020-03-31') 
data = pd.DataFrame({'Time': dates})  
print(data)
```

```python
for stock in stocks:
    prices = pd.read_csv("/home/ananya/Ananya/Portfolio_optmization/"+stock+".csv", usecols=['Date', 'Adj Close']
                        )
    #print(prices)
    prices['Date'] = pd.to_datetime(prices['Date'])  
    #print(prices)
    prices.rename(                                                            
        columns={"Date": "Time", "Adj Close": stock},
        inplace=True
    )
    
    data = pd.merge(data,prices)                                                               
    print(data)

```

```python
data = data[data['Time'].dt.weekday < 5] 
data = data.dropna(axis=0, how='all') 
data.head()
```

```python
p = data .drop(['Time'], axis=1).tail(1).to_numpy()
p
```

```python
r = data[(data['Time'].dt.weekday == 4)& (data['Time'] >= '2019-01-01')] \
    .drop(['Time'], axis=1) \
    .pct_change(fill_method='ffill')
r
```

```python
mu = r.mean().to_numpy()
sigma = r.cov().to_numpy()

print("The mean of the weekly return is-\n ",mu)
n=len(stocks)

print("The variance is:-\n",sigma)
```

```python
# This is for maximum return

from typing import List
from azure.quantum.optimization import Problem, ProblemType, Term

# variable 
def Return(mu,n)->Problem:
    terms=[]
    
    for i in range(0,n):
        
        terms.append(Term(c=-1*mu[i] , indices=[i]))
    print("Number of terms in return: ", len(terms))
    return terms
            
```

```python
def risk(sigma,n):
    terms=[]
    
    for i in range(0,n):
        for j in range(0,n):
            terms.append(Term(c= sigma[i][j] , indices=[i,j]))
    print("Number of terms in risk: ", len(terms))
    return terms
```

```python
terms=[]
t1 = Return(mu,n)
t2 = risk(sigma,n)
terms= t1+t2
problem = Problem(name="Portfolio optimization", problem_type=ProblemType.pubo, terms=terms)
```

```python
from azure.quantum.optimization import ParallelTempering        

#problem
solver = ParallelTempering(workspace)
solution = solver.optimize(problem)
```

```python

print(solution['configuration'])
result = solution['configuration']
print(type(result))


# xT Q x
# calculating risk
```

```python
risk=0
for i in range(0,100):
    for j in range(0,100):
        t = result[str(i)]* result[str(j)]*sigma[i][j]
        risk+=t

print("The risk for the selected portfolio is", risk)
```
