```python
import os                               
import numpy as np                    
import pandas as pd                     
import matplotlib.pyplot as plt       
```

```python
stocks = [stock.split('.')[0] for stock in sorted(os.listdir("/home/ananya/Ananya/Portfolio_optmization"))]
#stocks
print(stocks)
```

```python
from azure.quantum import Workspace
workspace = Workspace (
    subscription_id = "6e02aef9-2670-4418-92b8-7659cf5605d2",
    resource_group = "azurequantum",
    name = "workspace1",
    location = "japaneast"
)
```

```python
dates = pd.date_range('2000-01-01', '2020-03-31') 
data = pd.DataFrame({'Time': dates})  
print(data)
```

```python
for stock in stocks:
    prices = pd.read_csv("/home/ananya/Ananya/Portfolio_optmization/"+stock+".csv", usecols=['Date', 'Adj Close']
                        )
    #print(prices)
    prices['Date'] = pd.to_datetime(prices['Date'])  
    #print(prices)
    prices.rename(                                                            
        columns={"Date": "Time", "Adj Close": stock},
        inplace=True
    )
    
    data = pd.merge(data,prices)                                                               
    #print(data)

```

```python
data = data[data['Time'].dt.weekday < 5] 
data = data.dropna(axis=0, how='all') 
data.head()
```

```python
p = data .drop(['Time'], axis=1).tail(1).to_numpy()
p
```

```python
r = data[(data['Time'].dt.weekday == 4)& (data['Time'] >= '2019-01-01')] \
    .drop(['Time'], axis=1) \
    .pct_change(fill_method='ffill')
r
```

```python
mu = r.mean().to_numpy()
sigma = r.cov().to_numpy()

#print("The mean of the weekly return is-\n ",mu)
n=len(stocks)

#print("The variance is:-\n",sigma)
```

```python
# This is for maximum return

from typing import List
from azure.quantum.optimization import Problem, ProblemType, Term

# variable 
def objectiveFunction(mu)->Problem:
    terms=[]
    
    for i in range(0,n):
        
        terms.append(Term(c=-1*mu[i] , indices=[i]))
    print(len(terms))
    #print(terms)
    return Problem(name="Portfolio Optimization problem" , problem_type=ProblemType.pubo, terms=terms)
            
```

```python
from azure.quantum.optimization import ParallelTempering        
problem = objectiveFunction(mu)
#problem
solver = ParallelTempering(workspace)
solution = solver.optimize(problem)
print(solution['configuration'])
result = solution['configuration']
print(type(result))

```

```python
# xT Q x
# calculating risk
risk=0

for i in range(0,100):
    for j in range(0,100):
        t = result[str(i)]* result[str(j)]*sigma[i][j]
        risk+=t

print(risk)
```

```python

# For minimum risk
from typing import List
from azure.quantum.optimization import Problem, ProblemType, Term


def objectivefunction2(mu):
    terms=[]
    
    for i in range(0,100):
        for j in range(0,100):
            terms.append(Term(c= sigma[i][j] , indices=[i,j]))
        
    return Problem(name="Portfolio Optimization problem2" , problem_type=ProblemType.pubo, terms=terms)
    
            
problem2= objectivefunction2(mu)
```

```python
solver = ParallelTempering(workspace)
solution = solver.optimize(problem2)
print(solution['configuration'])
result = solution['configuration']
print(type(result))
```
