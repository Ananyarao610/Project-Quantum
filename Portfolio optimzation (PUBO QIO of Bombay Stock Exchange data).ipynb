```python
import os                               
import numpy as np                    
import pandas as pd                     
import matplotlib.pyplot as plt  
import math
```

```python
stocks = [stock.split('.')[0] for stock in sorted(os.listdir("/Users/sharique/Desktop/Bombay Stock Exchange Top 50 (last 20 years data)"))]
stocks = stocks[1:]
print((stocks))
```

```python
dates = pd.date_range('2022-05-1', '2022-07-01') 
data = pd.DataFrame({'Time': dates})  
print(data)
```

```python
for stock in stocks:
    prices = pd.read_csv("/Users/sharique/Desktop/Bombay Stock Exchange Top 50 (last 20 years data)/"+stock+".csv", usecols=['Date', 'WAP'])
    #print(prices)
    prices['Date'] = pd.to_datetime(prices['Date'], dayfirst = True)  
    #print(prices)
    prices.rename(                                                            
        columns={"Date": "Time", "WAP": stock},
        inplace=True
    )
    data = pd.merge(data,prices)
print(data)
```

```python
cp = data .drop(['Time'], axis=1).tail(1).to_numpy()
cp = cp[0]
cp
```

```python
r = data[(data['Time'] >= '2022-05-01')] \
    .drop(['Time'], axis=1) \
    .pct_change(fill_method='ffill')
r
```

```python
mu = r.mean().to_numpy()
sigma = r.cov().to_numpy()

print("The mean of the monthly return is-\n ",mu)
n=len(stocks)

print("The variance is:-\n",sigma)
```

```python
from azure.quantum import Workspace
workspace = Workspace (
    subscription_id = "e87e6b08-4913-4522-9206-ba18a56ee0fa",
    resource_group = "AzureQuantum",
    name = "MTCProjectQuantumSharique",
    location = "East US"
)
```

```python
from typing import List
from azure.quantum.optimization import Problem, ProblemType, Term

# variable 
def Return(mu,n)->Problem:
    terms=[]
    
    for i in range(0,n):
        
        terms.append(Term(c=-1*mu[i] , indices=[i]))
    print("Number of terms in return: ", len(terms))
    return terms
```

```python
def risk(sigma,n,risk_aversion):
    terms=[]
    
    for i in range(0,n):
        for j in range(0,n):
            terms.append(Term(c= sigma[i][j] * (risk_aversion/2) , indices=[i,j]))
    print("Number of terms in risk: ", len(terms))
    return terms
```

```python
from azure.quantum.optimization import Term, SlcTerm, GroupType

def budget_constraint(penalty_weight,n,b):
    terms=[]
    for i in range(0,n):
        terms.append(Term(c= 1 , indices=[i]))
    terms.append(Term(c=-b, indices=[]))
    slc = []
    slc.append(SlcTerm(
            terms,
            c=1
        ))
    return slc
```

```python
terms=[]
t1 = Return(mu,n)
t2 = risk(sigma,n,1)
terms= t1+ t2 
problem = Problem(name="Portfolio optimization", problem_type=ProblemType.pubo, terms=terms)
```

```python
from azure.quantum.optimization import ParallelTempering
import time

# Instantiate a solver to solve the problem. 
solver = ParallelTempering(workspace)

# Optimize the problem
print('Submitting problem...')
start = time.time()
solution = solver.optimize(problem)
time_elapsed = time.time() - start
print(f'Result in {time_elapsed} seconds: ')
```

```python
print(solution['configuration'])
result = solution['configuration']
#print(type(result))

selected_stocks=[]
prices_selected=[]
index_prices_selected= []

daily_prices = data.drop(['Time'], axis=1).to_numpy()
#print(daily_prices)

pmu = []
for i in range(0,34):
    for j in range(0,n):
        pmu.append(daily_prices[i][j].mean()) 
#print(pmu)

for i in result.keys():
    if result[i]:
        selected_stocks.append(stocks[int(i)])
        prices_selected.append(pmu[int(i)])
        index_prices_selected.append(int(i))
        
print("Selected stocks are -\n", selected_stocks)

print("Price of selected stocks are -\n", prices_selected)
```

```python
def find_risk():
    risk=0
    for i in range(0,n):
        for j in range(0,n):
            t = 0.5 * result[str(i)]* result[str(j)]*sigma[i][j]
            risk+=t
    return risk
```

```python
from numpy import random

def distributed_budget(B, arr, cp, prices_selected):
    shares=[]
    distribution = B/len(arr)   #1000 rupees for each stock
    #How many shares can be bought?

    for i in range(len(arr)):
        numberOfShares = math.floor(distribution/cp[index_prices_selected[i]])   
        shares.append(numberOfShares)
        print(arr[i], " : " ,shares[i]," shares")
    return shares
```

```python
no_of_shares = distributed_budget(200000, selected_stocks, cp, prices_selected)
#type(no_of_shares[0])
```

```python

def find_return():
    Return=0
    number_stocks = len(selected_stocks)
    for i in range(0,number_stocks):
        t = prices_selected[i]*number_stocks
        Return+=t
    return Return
```

```python
# Printing results

print("OPTIMIZSED PORTFOLIO : \n")
ab= distributed_budget(200000, selected_stocks, cp, prices_selected)
print("\n Constraints:")
print("\n Budget : ", 200000,"\n")
print(" Risk Aversion index of :" , 1)
print("\nReturns : ", find_return())
print("\nRisk : ", round(find_risk()*100,3) ,"%")
```
